```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y  = make_regression(n_samples=4,n_features=1,n_informative =1,n_targets=1,noise=80,random_state=13)
```

```python
X
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
#applying OLS
from sklearn.linear_model import LinearRegression
```

```python
reg= LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

### Below it searches for value b at which slope is 0 and it gradually reaches to actual b =26.159 (which we got
### from Linear regression model)

```python
#lets apply Gradient Descent
#assume m = 78.35
m = 78.35
b = -100
lr = 0.01
epochs = 100
for i in range(epochs):
    loss_slope = -2*np.sum(y-m*X.ravel()-b)
    b = b - (lr*loss_slope)
    y_pred = m * X + b
    plt.plot(X,y_pred)
plt.scatter(X,y,color='black')
plt.plot(X,reg.predict(X),color='black',label ="b = 26.159(intercept)")
plt.legend()
```
